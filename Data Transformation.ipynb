```python
import pandas as pd
import numpy as np
import string
```

```python
referee_list = pd.DataFrame(columns = ["referee"])
for year in range(2014,2019):
    referee = pd.read_csv("./" + str(year) + "/" + str(year) + "_match-list.csv").referee
    referee_list = referee_list.append(pd.DataFrame(referee, columns = ["referee"]))
```

```python
def strip_accents(string):
    ans = string
    replacements = {
        "ç": "c",
        "é": "e",
        "'": " "
    }
    for el in replacements.keys():
        ans = ans.replace(el, replacements[el])
    return ans
```

```python
referee_list.referee = referee_list.referee.str.lower().str.strip().apply(strip_accents).str.replace(" ", "_")
```

```python
referee_list = referee_list.drop_duplicates()
```

```python
referees = pd.DataFrame({"referee_name": referee_list.referee, "referee_id": range(1, referee_list.shape[0]+1)})
```

```python
referees.to_csv("referees.csv", index=False)
```
