```python
def strip_accents(string):
    ans = string
    replacements = {
        "ç": "c",
        "é": "e",
        "'": " "
    }
    for el in replacements.keys():
        ans = ans.replace(el, replacements[el])
    return ans
```

```python
referee_list.referee = referee_list.referee.str.lower().str.strip().apply(strip_accents).str.replace(" ", "_")
```

```python
def scout_parser(scout, col_names):
    scout_dict = ast.literal_eval(scout)
    scout_list = []
    for col in col_names:
        try:
            el = scout_dict[col]
        except:
            el = 0
        scout_list.append(el)
    return pd.DataFrame(columns = col_names, data = [scout_list]).iloc[0]
```

```python
def match_candidates(batch, player_name, threshold):
    candidates = pd.DataFrame(columns=["ratio", "index_num"])
    for i in range(0,batch.shape[0]):
        comp_a = player_name.lower()
        comp_b = batch.iloc[i].player_nickname.lower()
        ratio1 = fuzz.ratio(comp_a,comp_b)
        ratio2 = fuzz.partial_ratio(comp_a,comp_b)
        ratio3 = fuzz.token_set_ratio(comp_a,comp_b)
        ratio4 = 0
        ratio5 = 0
        ratio6 = 0
        try:
            comp_c = batch.iloc[i].player_name.lower()
            ratio4 = fuzz.ratio(comp_a,comp_c)
            ratio5 = fuzz.partial_ratio(comp_a,comp_c)
            ratio6 = fuzz.token_set_ratio(comp_a,comp_c)
        except:
            pass
        max_ratio = max(ratio1,ratio2,ratio3,ratio4,ratio5,ratio6)
        if max_ratio >= threshold:
            candidates = candidates.append(pd.DataFrame(data = [[max_ratio, i]], columns=["ratio", "index_num"]))
    return candidates
```

```python
def fuzzy_match(player_name, team_id, scouts, threshold):
    batch = scouts.query("team_id == @team_id")[["player_id", "player_name", "player_nickname"]]
    candidates = match_candidates(batch, player_name, threshold)
    num_candidates = candidates.shape[0]
    if num_candidates == 0:
        match_player_id = None
        match_player_name = None
        match_player_nickname = None
        match_ratio = None
    elif num_candidates == 1:
        match = candidates.iloc[0]
        i = match.index_num
        batch_match = batch.iloc[i]
        match_player_id = batch_match.player_id
        match_player_name = batch_match.player_name
        match_player_nickname = batch_match.player_nickname
        match_ratio = match.ratio
    else:
        match = candidates.sort_values(by="ratio", ascending = False)
        match_player_id = []
        match_player_name = []
        match_player_nickname = []
        match_ratio = []
        for j in range(0,match.shape[0]):
            i = match.iloc[j].index_num
            batch_match = batch.iloc[i]
            match_player_id.append(batch_match.player_id)
            match_player_name.append(batch_match.player_name)
            match_player_nickname.append(batch_match.player_nickname)
            match_ratio.append(match.iloc[j].ratio)
    return pd.DataFrame(data = [[match_player_id, match_player_name, match_player_nickname, match_ratio]], columns=["player_id", "match_name", "match_nickname", "match_ratio"]).iloc[0]
```

```python
def fill_not_found(row, scouts, db_match):
    if row.player_id != None:
        ans_player_id = row.player_id
        ans_player_name = row.match_name
        ans_player_nickname = row.match_nickname
        ans_ratio = row.match_ratio
    else:
        team_id = row.team_id
        used_player_ids = list(db_match.query("team_id == @team_id").player_id)
        candidates = scouts.query("team_id == @team_id and player_id not in @used_player_ids")[["player_id", "player_name", "player_nickname"]]
        ans_player_id = list(candidates.player_id)
        ans_player_name = list(candidates.player_name)
        ans_player_nickname = list(candidates.player_nickname)
        ans_ratio = None
    return pd.DataFrame(data = [[ans_player_id, ans_player_name, ans_player_nickname, ans_ratio]], columns=["player_id", "match_name", "match_nickname", "match_ratio"]).iloc[0]
```

```python
teams = pd.read_csv("./teams.tsv", sep="\t")
for year in range(2018,2019):
    for round_num in range(1,39):
        filename1 = "./" + str(year) + "/scouts/" + str(year) + "_round"
        filename2 = "./" + str(year) + "/substitutions/" + str(year) + "_round"
        filename3 = "./" + str(year) + "/substitutions/" + str(year) + "_round"
        if round_num < 10:
            filename1 = filename1 + "0" + str(round_num) + "_scouts.csv"
            filename2 = filename2 + "0" + str(round_num) + "_substitutions.csv"
            filename3 = filename3 + "0" + str(round_num) + "_substitution_report.csv"
        else:
            filename1 = filename1 + str(round_num) + "_scouts.csv"
            filename2 = filename2 + str(round_num) + "_substitutions.csv"
            filename3 = filename3 + str(round_num) + "_substitution_report.csv"
        try:
            scouts = pd.read_csv(filename1)
            substitutions = pd.read_csv(filename2)
            db_match = pd.merge(substitutions, teams, on = "team_name_3", how = "left").iloc[:,:5]
            db_match[["player_id", "match_name", "match_nickname", "match_ratio"]] = db_match.apply(lambda row: fuzzy_match(row.player_name_2, row.team_id, scouts, 80), axis = 1)
            substitution_report = db_match.copy()
            substitution_report[["player_id", "match_name", "match_nickname", "match_ratio"]] = db_match.apply(lambda row: fill_not_found(row, scouts, db_match), axis=1)
            substitution_report.to_csv(filename3, index=False)
        except Exception as e:
            print(e, year, round_num)
```
