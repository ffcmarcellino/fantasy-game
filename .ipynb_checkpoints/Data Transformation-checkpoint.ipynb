```python
import pandas as pd
import numpy as np
import string
from datetime import datetime as dt
from datetime import timedelta as td
```

```python
referee_list = pd.DataFrame(columns = ["referee"])
for year in range(2014,2019):
    referee = pd.read_csv("./" + str(year) + "/" + str(year) + "_match-list.csv").referee
    referee_list = referee_list.append(pd.DataFrame(referee, columns = ["referee"]))
```

```python
def strip_accents(string):
    ans = string
    replacements = {
        "ç": "c",
        "é": "e",
        "'": " "
    }
    for el in replacements.keys():
        ans = ans.replace(el, replacements[el])
    return ans
```

```python
referee_list.referee = referee_list.referee.str.lower().str.strip().apply(strip_accents).str.replace(" ", "_")
```

```python
referee_list = referee_list.drop_duplicates()
```

```python
referees = pd.DataFrame({"referee_name": referee_list.referee, "referee_id": range(1, referee_list.shape[0]+1)})
```

```python
referees.to_csv("referees.csv", index=False)
```

```python
team_list = pd.DataFrame(columns = ["team_name_3"])
for year in range(2014,2019):
    filename = "./" + str(year) + "/" + str(year) + "_match-list.csv"
    match_list = pd.read_csv(filename)
    match_list.team_home = match_list.team_home.str.replace("Atletico - PR", "Atlético - PR")
    match_list.team_away = match_list.team_away.str.replace("Atletico - PR", "Atlético - PR")
    match_list.to_csv(filename, index=False)
```

```python
for round_num in range(1,39):
    filename = "./2018/scouts/2018_round"
    if round_num < 10:
        filename = filename + "0" + str(round_num) + "_scouts.csv"
    else:
        filename = filename + str(round_num) + "_scouts.csv"
    scouts = pd.read_csv(filename)
    scouts.team_id = scouts.apply(lambda row: 354 if row.team_id == '204/354' else row.team_id, axis = 1)
    scouts.to_csv(filename, index = False)
```

```python
old_mode = "Domingo"
d = dt(2018,4,10,14)
days = None
for round_num in range(1,39):
    filename = "./history/2018/rodada-" + str(round_num) + ".csv"
    scouts = pd.read_csv(filename)
    status = scouts[["atletas.rodada_id", "atletas.atleta_id", "atletas.status_id"]]
    status["query_time"] = d.isoformat(sep=" ")
    status.columns = ["round_num", "player_id", "status_id", "query_time"]
    status = status[["round_num", "query_time", "player_id", "status_id"]]
    filename2 = "./2018/status/2018_round"
    if round_num < 10:
        filename2 = filename2 + "0" + str(round_num) + "_status.csv"
    else:
        filename2 = filename2 + str(round_num) + "_status.csv"
    status.to_csv(filename2, index=False)
    
    match_list = pd.read_csv("./2018/2018_match-list.csv").query("round_num == @round_num")
    new_mode = match_list.groupby("week_day").match_num.count().sort_values(ascending=False).reset_index().week_day[0]
    if old_mode in ["Sábado", "Domingo"]:
        if new_mode in ["Sábado", "Domingo"]:
            days = 7
        else:
            days = 3
    else:
        if new_mode in ["Sábado", "Domingo"]:
            days = 4
        else:
            days = 7
    d = d + td(days)
    old_mode = new_mode
```
