```python
import pandas as pd
import math
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
from selenium.common.exceptions import TimeoutException
```

```python
main_url = "https://www.cbf.com.br/futebol-brasileiro/competicoes/campeonato-brasileiro-serie-a/"
option = webdriver.ChromeOptions()
option.add_argument(" â€” incognito")
browser = webdriver.Chrome(executable_path="./chromedriver", chrome_options=option)
```

```python
col_names = ["round_num", "team_home", "team_away", "week_day", "time", "stadium", "referee"]

for year in range(2018,2019):
    print(year)
    url_year = main_url + str(year)
    match_list = pd.DataFrame(columns= col_names)
    file_name = './' + str(year) + "/match-list/" + str(year) + '_match-list.csv'
    for i in range(1, 381):
        print(i)
        url = url_year + '/'+ str(i)
        
        # BeautifulSoup Webscrapping
        page = requests.get(url)
        parsed_page = bs(page.content, 'html.parser')
        ## Get team names
        teams = parsed_page.find_all('h3', class_='time-nome color-white')
        team_home = teams[0].get_text()
        team_away = teams[1].get_text()
        ## Get stadium name and match schedule
        info = parsed_page.find_all('span', class_='text-2 p-r-20')
        stadium = info[0].get_text().strip()
        week_day = info[1].get_text().split(",")[0].strip()
        time = info[2].get_text().strip()
        
        # Selenium Webscrapping
        browser.get(url)
        browser.find_elements_by_xpath("//a[@href='#arbitros']")[0].click()
        ## Get referee name
        referee = browser.find_element_by_xpath("//section[@id='#arbitros']").find_element_by_xpath("//td").text
        
        # Save data
        round_num = str(math.ceil(i/10))
        match = pd.DataFrame([[round_num, team_home, team_away, week_day, time, stadium, referee]], columns = col_names) 
        match_list = match_list.append(match, ignore_index = True)
    match_list.to_csv(file_name)
```
